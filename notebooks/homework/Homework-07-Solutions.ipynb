# Plotting and Interpretting Table Data

In this exercise, you will work with some real data obtained from an experiment with a California sea hare (*Aplysia californica*). The animal was fed 4 different types of food:
* **Natural, live, "fresh" seaweed**;
* **Regular strips of nori**, which are lightweight, delicate, and very easy to swallow;
* **Tape nori**, which is unbreakable double-sided tape (plastic) sandwiched between nori strips, and which was anchored to an immovable force transducer; and
* **Two-ply nori**, which are multiple layers of nori bound together and cut into wide strips, making them very difficult to swallow.

> seaweeds.jpg

The animal was also instrumented with implanted electrodes, so that the activity of its nervous system could be monitored while it fed. When feeding on tape nori, the amount of pulling force generated by the animal was also recorded.

Below is an example of the raw data obtained during the experiment, showing 4 swallows on tape nori. The bottom trace shows that the force rose and fell as the animal rhythmically pulled on the food with its feeding grasper. Each colored trace plots the electrical activity of a muscle or nerve. The yellow trace shows activity of **buccal nerve 2 (BN2), which carries motor signals that cause the feeding grasper to pull food into the animal**. The main burst of motor neuronal activity during each of the 4 swallows is enclosed in a box.

> recordings.png

More than one hundred swallows performed by this animal have already been analyzed. For each swallow, the following was measured:

* The **duration of the BN2 burst** (length of the boxes shown above);
* The **mean rectified voltage of the BN2 burst**, which is a measure of the overall level or intensity of activity during the burst (average of the absolute value of the signal inside the boxes); and
* The **peak force**, if the food type was tape nori (otherwise, no force measurement was taken).

These measurements for the 100+ swallows, along with the type of food the animal was eating at the time, have been organized into a spreasheet, which you can view [here](https://github.com/jpgill86/python-for-neuroscientists/blob/master/notebooks/homework/data/JG08-foods-BN2-force.csv). Note that the "Force peak" column is missing many values. This is because force measurements were not taken when the animal fed on foods other than tape nori.

For this exercise, you will import this spreadsheet into Python and create a couple plots to help make sense of the data.

## Part 2: Plot force vs BN2 activity

Using pandas and/or seaborn, create a scatter plot with
* **peak force on the vertical axis**, and
* **mean rectified voltage on the horizontal axis**.

Does the data suggest a relationship between buccal nerve 2 activity and swallowing force?

---
### **SOLUTION:**

**Data interpretation:** Peak swallowing force increases with intensity of BN2 activity, suggesting the BN2 motor neurons are recruited to generate more force.

````
# using a pandas plotting functions
df.plot.scatter(
    x='BN2 burst mean rectified voltage (μV)',
    y='Force peak (mN)',
    )
````

````
import seaborn as sns

# equivalent plot made with seaborn
sns.scatterplot(
    data=df,
    x='BN2 burst mean rectified voltage (μV)',
    y='Force peak (mN)',
    )
````

## Part 3: Plot BN2 activity for different foods

Using seaborn, create a scatter plot with
* **duration of the BN2 burst on the vertical axis**, and
* **mean rectified voltage on the horizontal axis**, with
* swallows on **each food type plotted in a different color**.

What differences in BN2 activity do you observe between the food types? How could you explain these differences in terms of the mechanical properties of the food and the nervous system of the animal?

---
### **SOLUTION:**

**Data interpretation:** Regular nori, tape nori, and two-ply nori occupy distinct spaces in the plot. The bottom-left corner, occupied by lightweight and delicate regular nori, represents minimal effort on the part of the animal: short duration and low intensity of motor neuronal activity. Both tape nori and two-ply nori are more difficult to swallow and required longer duration and higher intensity of motor neuronal activity. In contrast to these uniform strips, natural food is more complex and so posed a varied challenge to the animal, requiring it to vary its level of effort, resulting in the data points for fresh food being scattered all over the plot.

````
import seaborn as sns
````

````
# version 1 using scatterplot

sns.scatterplot(
    data=df,
    x='BN2 burst mean rectified voltage (μV)',
    y='BN2 burst duration (s)',
    hue='Food',
    )
````

````
# version 2 using relplot

sns.relplot(
    data=df,
    kind='scatter',
    x='BN2 burst mean rectified voltage (μV)',
    y='BN2 burst duration (s)',
    hue='Food',
    )
````
